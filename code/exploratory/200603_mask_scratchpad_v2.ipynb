```python
def remove_duplicate_rows(data):
    "Remove duplicated rows from a numpy array ."
    sorted_data =  data[np.lexsort(data.T),:]
    row_mask = np.append([True],np.any(np.diff(sorted_data,axis=0),1))
    return sorted_data[row_mask]

def get_xy_coords_mask(fname):
    "Helper function to get xy coordinates from .h5 mask file"

    # Read h5 file and convert to numpy array
    mask_h5 = np.array(h5py.File(fname, 'r')['data'])

    # Get xyz coordinates and remove duplicate rows
    xyz_coords = remove_duplicate_rows(np.transpose(np.nonzero(mask_h5)))
    
    # Get xy coords, note that order is scrambled 
    x_coords = xyz_coords[:, 1]
    y_coords = xyz_coords[:, 0]

    return x_coords, y_coords
```

```python
def get_reference_brain_zslice(path_to_mat, slice_ix = 25, return_df = False): 

    """
    Helper function to extract an anatomical slice in the z-axis from the fish reference brain 

    Params
    --------

    path_to_mat(str)
        Path to the ReferenceBrain.mat file.

    data_path (str)
        Path to store the dataset.

    slice_ix(int)
        Index of the slice to get.

    return_df (bool)
        If set to True it will return the dataframe.

    """

    anat = sio.loadmat(path_to_mat)

    anat_stack_norm = np.array(anat['anat_stack_norm'])

    zslice = anat_stack_norm[:, :, slice_ix]

    return zslice 
```
