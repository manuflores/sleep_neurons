```python
# Load data
def cluster_data(path):
    neuron_data = sl.load_ahrens_data_figshare(path)

    report_df = sl.cluster_neuron_data_v2(
        neuron_data, pca_components=0.6, n_neighbors=30, max_clus=40
    )
    
    return report_df 
```

```python
for path_to_data in dsets:

    fish_numba = dsets[0].split("/")[-2].split("_")[-1]

    report_df = cluster_data(path_to_data)

    report_df.to_csv(
        path_clus + "200501_report_df_clus_fish" + str(fish_numba) + ".csv", index=False
    )
```
